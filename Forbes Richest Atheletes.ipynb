This dataset consists of a complete list of the world’s highest-paid athletes since Forbes's first list in 1990. In 2002, they changed the reporting period from the full calendar year to June-to-June, and consequently, there are no records for 2001. The dataset consists of records till the year 2021.

# Introduction to Dataset

* Name: Contains Name of Athletes whose name appeared in Forbes list
* Nationality: Contains Nationality of the Athletes
* Earnings: Annual Earnings of Athletes
* Year: Year
* Sport: Sport affiliated to the Athlete

# 3. Exploratory Data Analysis

## 3.1 What is the distribution of earnings among athletes?

* Create a histogram or box plot to visualize the distribution of earnings.

```python
# Plot a histogram of earnings

plt.figure(figsize = (12,6))
plt.hist(x = df["Earnings"], bins=(20),align='mid',color='skyblue', edgecolor='black')
plt.title("Distribution of Earnings Among Athletes")
plt.xlabel('Earnings')
plt.ylabel('Frequency')
plt.show()
```

* Observation: The histogram  of earnings shows a skewed distribution, indicating that the majority of athletes have relatively low earnings, with a few high-earning outliers.

## 3.2 How have earnings in sports evolved over the years?

* Create a line plot to show the trend of earnings over different years.

```python
yearly_earnings = df.groupby("Year")["Earnings"].mean()
# Create a box plot of earnings
plt.figure(figsize = (12,6))
sns.lineplot(yearly_earnings,marker='o', linestyle='-')
plt.title("Trend of Earnings in Sports Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Earnings")
plt.show()
```

* Observation: The line plot of earnings over different years reveals trends in earnings across time. It may show increasing or decreasing patterns or periods of stability in earnings for athletes.

## 3.3 Which sport has the highest average earnings?

* Create a bar chart to compare the average earnings across different sports.

```python
average_earnings_by_sport = df.groupby("Sport")["Earnings"].mean()
average_earnings_by_sport = average_earnings_by_sport.sort_values(ascending=False)
# Create a box bar of earnings
plt.figure(figsize = (12,6))
plt.bar(average_earnings_by_sport.index, average_earnings_by_sport.values, color='skyblue')
plt.title('Average Earnings by Sport')
plt.xlabel("Sport")
plt.ylabel("Average Earnings")
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.show()
```

* Observation: The bar chart  displays the sport(s) with the highest average earnings. It identifies the sport(s) that tend to generate the most income for athletes.

## 3.4 What is the overall distribution of sports in the dataset?

* Create a bar chart or pie chart to show the frequency of each sport.

```python
# Count the frequency of each sport
sport_counts = df['Sport'].value_counts()

# Create a bar chart to show the frequency of each sport
plt.figure(figsize=(10, 6))
plt.bar(sport_counts.index, sport_counts.values, color='skyblue')
plt.title('Frequency of Sports in the Dataset')
plt.xlabel('Sport')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

* Observation: The bar chart illustrates the frequency of each sport in the dataset, providing an overview of the diversity or concentration of sports.

## 3.5 Is there a relationship between nationality and earnings?

* Create a scatter plot or box plot to visualize the relationship between nationality and earnings.

```python
# Create a scatter plot to visualize the relationship between nationality and earnings
plt.figure(figsize=(12, 6))
plt.scatter(df['Nationality'], df['Earnings'], alpha=0.5, color='skyblue')
plt.title('Relationship Between Nationality and Earnings')
plt.xlabel('Nationality')
plt.ylabel('Earnings')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

* Observation: The scatter plot reveals whether there is a discernible relationship between an athlete's nationality and their earnings. It can help identify patterns or disparities.

## 3.6 Are there any outliers in earnings?

* Create a box plot to identify potential outliers in earnings.

```python
# Create a box plot to identify potential outliers in earnings
plt.figure(figsize=(8, 6))
plt.boxplot(df['Earnings'], vert=False)
plt.title('Box Plot of Earnings to Identify Outliers')
plt.xlabel('Earnings')
plt.tight_layout()
plt.show()
```

* Observation: The box plot identifies potential outliers in earnings, helping to pinpoint extreme values that deviate significantly from the typical earnings distribution.

## 3.7 Which nationality has the highest number of athletes in the dataset?

* Create a bar chart to display the frequency of each nationality.

```python
# Count the frequency of each nationality
nationality_counts = df['Nationality'].value_counts()

# Create a bar chart to display the frequency of each nationality
plt.figure(figsize=(12, 6))
plt.bar(nationality_counts.index, nationality_counts.values, color='skyblue')
plt.title('Frequency of Athletes by Nationality')
plt.xlabel('Nationality')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

* Observation: The bar chart highlights the nationality with the highest representation among athletes, indicating which nationality is the most common in the dataset.

## 3.8 How does the distribution of earnings vary for different sports?

* Create separate histograms or box plots for each sport to compare earnings distributions.

```python
# Create separate box plots for each sport
plt.figure(figsize=(12, 6))
boxplot_data = [df[df['Sport'] == sport]['Earnings'] for sport in df['Sport'].unique()]
plt.boxplot(boxplot_data, labels=df['Sport'].unique(), vert=False)
plt.title('Distribution of Earnings by Sport (Box Plots)')
plt.xlabel('Earnings')
plt.tight_layout()
plt.show()
```

* Observation: box plots for each sport show how earnings distributions vary across different sports, helping to identify disparities or similarities.

## 3.9 Are there any trends or patterns in earnings over the years for specific sports or nationalities?

* Create line plots or time series plots for specific sports or nationalities to identify trends.

```python
# Select a specific sport for analysis
selected_sport = 'Football'  # Replace with the sport you're interested in

# Filter the data for the selected sport
sport_data = df[df['Sport'] == selected_sport]

# Group the data by year and calculate the mean earnings for each year
yearly_earnings = sport_data.groupby('Year')['Earnings'].mean()

# Create a line plot to show earnings trends for the selected sport
plt.figure(figsize=(12, 6))
plt.plot(yearly_earnings.index, yearly_earnings.values, marker='o', linestyle='-')
plt.title(f'Earnings Trend Over the Years for {selected_sport}')
plt.xlabel('Year')
plt.ylabel('Average Earnings')
plt.grid(True)
plt.show()
```

* Observation: Line plots  for specific sports or nationalities reveal trends, patterns, or anomalies in earnings over time, providing insights into how economic factors impact athletes.

## 3.10 What are the top-earning athletes, and which sports are they associated with?

* Create a table or bar chart to display the top-earning athletes and their respective sports.

```python
# Sort the DataFrame by earnings in descending order to get the top earners
top_earning_athletes = df.sort_values(by='Earnings', ascending=False).head(10)

# Create a bar chart to display the top-earning athletes and their respective sports
plt.figure(figsize=(12, 6))
plt.barh(top_earning_athletes['Name'], top_earning_athletes['Earnings'], color='skyblue')
plt.title('Top-Earning Athletes and Their Respective Sports')
plt.xlabel('Earnings')
plt.ylabel('Athlete')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest earner at the top
plt.tight_layout()
plt.show()
```

* Observation: The  bar chart displays the names of the top-earning athletes along with their respective sports, allowing you to identify the highest-earning individuals and their disciplines.
